```python
import cv2
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,720)
```

```python
%%time
ret, frame = cap.read()
```

```python
%%time
cap.grab()
```

```python
%%time
ret, frame = cap.retrieve()
```

```python
plt.imshow(frame);
```

```python
cap.release()
```
