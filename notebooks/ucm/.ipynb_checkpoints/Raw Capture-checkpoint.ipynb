```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
cap = cv2.VideoCapture(1)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,720)
```

```python
ret, frame = cap.read()
```

```python
matplotlib.rcParams['figure.figsize'] = [18, 9]
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB));
```

```python
cap.release()
```

```python
(rows,cols,channels) = frame.shape
image_size = cols / 2
front_img = frame[0:image_size, 0:image_size]
back_img  = frame[0:image_size, image_size:2*image_size]
```

```python
matplotlib.rcParams['figure.figsize'] = [18, 9]
plt.subplot(121); plt.imshow(cv2.cvtColor(front_img, cv2.COLOR_BGR2RGB));
plt.subplot(122); plt.imshow(cv2.cvtColor(back_img, cv2.COLOR_BGR2RGB));
```

```python
cv2.imwrite("front.png", front_img)
cv2.imwrite("back.png", back_img)
```
