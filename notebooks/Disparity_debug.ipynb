```python
valid = (disparity>0)
```

```python
invalid = (disparity==-1)
```

```python
plt.imshow(invalid,cmap='gray');
```

```python
sinvnfs = np.zeros((640,1280), dtype=np.float32)
for u in range(1280):
    sinvnfs[:,u] = np.sin(np.arange(640, dtype=np.float32) * np.pi / 640)
```

```python
distance = 0.35 * sinvnfs / np.sin(disparity*np.pi/640.0)
```

```python
distance = np.clip(distance, 0, 20)
```

```python
plt.imshow(distance*valid + 20*invalid,cmap='gray');
```
