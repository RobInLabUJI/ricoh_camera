```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
p_tbf = np.load('p_tbf.npy')

tpt = p_tbf[:,0:2]
bpt = p_tbf[:,2:4]

matplotlib.rcParams['figure.figsize'] = [15, 15]
plt.subplot(121); plt.imshow(plt.imread("Top BF stitch_screenshot_13.06.2019.png"));
plt.subplot(122);
plt.scatter(tpt[:,0], tpt[:,1], c='red'); plt.scatter(bpt[:,0]+320, bpt[:,1], c='green');
for m in p_tbf[::10,:]:
    plt.plot([m[0],m[2]+320],[m[1],m[3]])
plt.axis('scaled'); plt.axis([0, 640, 0, 640]);  plt.gca().invert_yaxis();
```

```python
th = np.arctan2(bpt[:,1]-tpt[:,1], bpt[:,0]+320-tpt[:,0])

matplotlib.rcParams['figure.figsize'] = [7, 7]
plt.hist(th, len(th)/20);
```

```python
idx = np.logical_and(th>-0.18, th<-0.05)
p_tbf = np.load('p_tbf.npy')[idx,:]

tpt = p_tbf[:,0:2]
bpt = p_tbf[:,2:4]

matplotlib.rcParams['figure.figsize'] = [15, 15]
plt.subplot(121); plt.imshow(plt.imread("Top BF stitch_screenshot_13.06.2019.png"));
plt.subplot(122);
plt.scatter(tpt[:,0], tpt[:,1], c='red'); plt.scatter(bpt[:,0]+320, bpt[:,1], c='green');
for m in p_tbf[::10,:]:
    plt.plot([m[0],m[2]+320],[m[1],m[3]])
plt.axis('scaled'); plt.axis([0, 640, 0, 640]);  plt.gca().invert_yaxis();
```
