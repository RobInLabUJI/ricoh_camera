```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
p_tbf = np.load('p_tbf.npy')

tpt = p_tbf[:,0:2]
bpt = p_tbf[:,2:4]

matplotlib.rcParams['figure.figsize'] = [15, 15]
plt.subplot(121); plt.imshow(plt.imread("Top BF stitch_screenshot_13.06.2019.png"));
plt.subplot(122); plt.scatter(tpt[::10,0], tpt[::10,1], c='red'); plt.scatter(bpt[::10,0]+320, bpt[::10,1], c='green');
for m in p_tbf[::10,:]:
    plt.plot([m[0],m[2]+320],[m[1],m[3]])
plt.axis('scaled'); plt.axis([0, 640, 0, 640]);  plt.gca().invert_yaxis();
```

```python
dy = tpt[:,1]-bpt[:,1]
matplotlib.rcParams['figure.figsize'] = [7, 7]
n, b, patches = plt.hist(dy, len(dy)/20);
bin_max = np.where(n == n.max())
print 'maxbin', b[bin_max][0]
```

```python
idx = np.logical_and(dy>25, dy<50)
p_tbf = np.load('p_tbf.npy')[idx,:]

tpt = p_tbf[:,0:2]
bpt = p_tbf[:,2:4]

matplotlib.rcParams['figure.figsize'] = [15, 15]
plt.subplot(121); plt.imshow(plt.imread("Top BF stitch_screenshot_13.06.2019.png"));
plt.subplot(122); plt.scatter(tpt[:,0], tpt[:,1], c='red'); plt.scatter(bpt[:,0]+320, bpt[:,1], c='green');
for m in p_tbf[::,:]:
    plt.plot([m[0],m[2]+320],[m[1],m[3]])
plt.axis('scaled'); plt.axis([0, 640, 0, 640]);  plt.gca().invert_yaxis();
```

```python
idx = tpt[:,0]>=160 # top_f camera only
tptf = tpt[idx,:]
bptf = bpt[idx,:]

idx = tpt[:,0]<160 # top_b camera only
tptb = tpt[idx,:]
bptb = bpt[idx,:]

print np.median(bptf[:,0] -  tptf[:,0]), len(tptf)
print np.median(bptb[:,0] -  tptb[:,0]), len(tptb)
```
