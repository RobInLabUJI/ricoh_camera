```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [15, 15]
```

```python
p_tbf = np.load('p_tbf.npy')
p_tfb = np.load('p_tfb.npy')
p_bbf = np.load('p_bbf.npy')
p_bfb = np.load('p_bfb.npy')
```

```python
tpt = p_tbf[:,0:2]
bpt = p_tbf[:,2:4]

plt.subplot(121); plt.imshow(plt.imread("Top BF stitch_screenshot_13.06.2019.png"));
plt.subplot(122);
plt.scatter(tpt[:,0], tpt[:,1], c='red'); plt.scatter(bpt[:,0]+320, bpt[:,1], c='green');
for m in p_tbf[::10,:]:
    plt.plot([m[0],m[2]+320],[m[1],m[3]])
plt.axis('scaled'); plt.axis([0, 640, 0, 640]);  plt.gca().invert_yaxis();
```
